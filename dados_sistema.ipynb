```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("dados_turmas\Relatório de Rendimento Escolar_900.xlsx", skiprows=12, engine="openpyxl", index_col=0)
df.head(3)
```

```python
df = df.drop(df.index[-2:])
df.head()
```

```python
df.drop(columns=df.columns[12:], inplace=True)
df.head(2)
```

```python
colunas_rename = ["Nome","Notas_1bim","Faltas_1bim","Notas_2bim","Faltas_2bim","Notas_3bim","Faltas_3bim","Notas_4bim","Faltas_4bim","Rec","Media","Faltas"]

df.columns = colunas_rename
df.drop(columns=["Rec","Media","Faltas"], inplace=True)
df.head()
```

```python
df_long = df.melt(
    id_vars=["Nome"],
    var_name="tipo",
    value_name="valor"
)

df_long[['variavel','bimestre']] = df_long['tipo'].str.split("_", expand=True)

df_long['bimestre'] = df_long['bimestre'].str.extract('(\d+)').astype(int)

df_final = df_long.pivot_table(
    index=["Nome","bimestre"],
    columns = "variavel",
    values = "valor",
    aggfunc = "first").reset_index()

df_final.columns.name = None

df_final = df_final.rename(columns={'nota': 'notas', 'faltas': 'faltas'})

df_final = df_final.sort_values(by=['bimestre', 'Nome']).reset_index(drop=True)

df_final.head(30)
```
